```python
# read hetero/noise_vs_height_va.csv with delimiter: ,
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# read csv file
df_va = pd.read_csv('hetero/noise_vs_height_va.csv')
df_dva = pd.read_csv('hetero/noise_vs_height_dva.csv')

# plot  
plt.figure(figsize=(10, 5))
start, end = 100, -50
plt.plot(df_va['w'][start:end], df_va['true_std'][start:end], label='true')
plt.plot(df_va['w'][start:end], df_va['va_sigma'][start:end], label='VA')
plt.plot(df_dva['w'][start:end], df_dva['dva_sigma'][start:end], label='DVA')
plt.xlabel('Height')
plt.ylabel('Std')
plt.title('Noise at a = 0.1')
plt.legend()
```

```python
# read all pickle files in results/noise_1.0
# and save them in a dataframe

import os
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# read pickle files
path = 'results/noise_0.1'
files = os.listdir(path)
files = [f for f in files if f.endswith('.pkl')]
files.sort()

# read pickle files and save them in a dataframe
df = pd.DataFrame()
for f in files:
    with open(path + '/' + f, 'rb') as file:
        df_temp = pickle.load(file)
        for key, val in df_temp.items():
            print(key, val)
        # show all values as a comma dilimited list, and break out the list into columns
        print(','.join([str(val) for val in df_temp.values()]))
```
